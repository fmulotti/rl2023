```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bras1_moyenne=0.5
bras1_et=0.10

bras2_moyenne=0.45
bras2_et=0.20
```

```python
sample = [ 10, 100, 1000, 1000, 10000 ]
for i in sample:
  np.random.seed(2024)
  bras1=np.random.normal(0.5,0.10,i)
  bras2=np.random.normal(0.4,0.3,i)
  plt.figure()
  plt.hist(bras2,color="RED",label="Bras2",bins=51)
  plt.hist(bras1,color="blue",label="bras1",bins=51)
  plt.legend()
  plt.xlim(0,1)
  plt.ylabel("Fr√©quence")
  plt.title(f"{i} tirages")
  plt.show()
```

```python
print(f"Moyenne bras 1 : {np.average(bras1)}, moyenne bras 2 : {np.average(bras2)}")
```
